## 1. Matplotlib

Matplotlib is an excellent 2D and 3D graphics library for generating scientific figures. One of the key features of matplotlib is that all aspects of the figure can be controlled *programmatically*. This is important for reproducibility and convenient when one needs to regenerate the figure with updated data or change its appearance. 

More information at the Matplotlib web page: http://matplotlib.org/

```python
# import the required packages
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
from scipy.stats import norm
```

### 1.1 line graph
#### 1.1.1 Draw a line

```python
x = np.arange(-3, 3, 0.01)
# Generate a Bell Curve (Probability Density Function for a normally distribued random variable with mean 0 and std 1)
y = norm.pdf(x) 

plt.plot(x, y)
plt.show()
```

```python
y
```

#### 1.1.2 Mutiple Lines

```python
plt.plot(x, norm.pdf(x)) # line 1: a bell curve with mean 0 and std 1
plt.plot(x, norm.pdf(x, 1.0, 0.5)) # Line 2: a bell curve with mean 1 and std 0.5
plt.show()
```

#### 1.1.3 Save it to a File

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.savefig('MyPlot.png', format='png')
```

#### 1.1.4 Adjust the Axes

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.axis([-5, 5, 0, 1]) # [Xmin, Xmax, Ymin, Ymax]
plt.show()
```

#### 1.1.5 Add a Grid

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.axis([-5, 5, 0, 1])
plt.grid()
plt.show()
```

#### 1.1.6 Change Line Types and Colors

```python
plt.plot(x, norm.pdf(x), 'b-') # Line 1: Blue Color Line
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:') # Line 2: Red Color dotted-line
plt.axis([-5, 5, 0, 1])
plt.grid()
plt.show()
```

#### 1.1.7 Labeling Axes and Adding a Legend

```python
plt.xlabel('X-Label')
plt.ylabel('Y-label')
plt.plot(x, norm.pdf(x), 'b-')
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:')
plt.axis([-5, 5, 0, 1])
plt.grid()
plt.legend(['Line 1', 'Line 2'], loc=1)
plt.title('Title')
plt.show()
```

### 1.6 Box & Whisker Plot

Useful for visualizing the spread & skew of data.

```python
plt.boxplot(age)
plt.title('Box Plot of Age')
plt.show()
```

* The red line represents the median of the data (i.e. `Q2 = 40`)
* The box represents the bounds of the 1st quartile (i.e. `Q1 = 33`) and 3rd quartiles (i.e. `Q3 = 47`)
* The Interquartile Range (IQR) is the distance from Q1 to Q3 (i.e. `IQR = Q3 - Q1 = 14`)
* The "whiskers" indicate the range of the data except outliers, which are 1.5 times from the Interquartile Range (IQR). To be specific, the whiskers "Minimum" is `Q1 - 1.5 * IQR = 12` and the whiskers "Maximum" is `Q3 + 1.5 * IQR = 68`.
* The outliers are plotted outside of the whiskers (i.e. smaller than 12 or bigger than 68).
