```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
house = pd.read_csv('./Mumbai.csv', encoding='utf-8')
house
```

```python
house = house.drop(columns=['Location'])
house
```

```python
house.info()
```

```python
house.isnull().sum()
```

```python
house.describe()
```

```python
plt.figure(figsize=(50,50))
ax = sns.heatmap(house.corr(), annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

```python
house_corr = house.iloc[:,0:3]
plt.figure(figsize=(10,10))
ax = sns.heatmap(house_corr.corr(), annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

```python
plt.boxplot(x=house.iloc[:,0:3])
plt.show()
```

```python
plt.hist(house['Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Counts') 
plt.grid(True)
plt.show()
```

```python
plt.scatter(house['Area'], house['Price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area Vs Price')
plt.show()
```
